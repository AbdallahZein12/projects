## Terms to Know for Audio in Digital Form

### Frequency (Hz)

> - Frequency describes the differences of wave lengths  
> - We interperate frequency as high and low pitches.  

  
<img src='https://www.physicsclassroom.com/Class/sound/u11l2a2.gif'>

### Intensity (db/power)
> - Intensity describes the amplitude (height) of the wave.

<img src='https://ars.els-cdn.com/content/image/3-s2.0-B9780124722804500162-f13-15-9780124722804.gif'>

### Sample Rate 

> - Sample rate is specific to how the computer reads in the audio file.
> - Think of it as the "resolution" of the audio

<img src='https://www.ooberpad.com/cdn/shop/articles/Sampling_Rate_Featured_image.jpg?v=1654170297'>

### Reading in Audio Files
Audio types can include: `mp3`, `wav`, `m4a`, `flac`, `ogg`

```python
audio_files = glob('./Dataset/*/*.wav')
```

```python
# Play Audio File
ipd.Audio(audio_files[0])
```

```python
y, sr = librosa.load(audio_files[0])
print(f"y: {y[:10]}\n\nshape y: {y.shape}\n\nsr: {sr}")
```

```python
pd.Series(y).plot(figsize=(10,5),lw=1,title='Raw Audio Example',color=color_pal[0])
plt.show()
```
